```python

def expo(lam,n):

  Mu=1/lam
  L=-Mu
  C=[]
  for i in range(n):

      x=rd.random()
      U=L*np.log10(x)
      C.append(U)


  return C
```

```python
def cdf(muestra):

  n=len(muestra)
  x=np.sort(muestra)
  y = np.arange(1, n + 1) / n
  return x,y
```

```python

def Cauchy(gam,x0,n):
  C=[]
  for i in range(n):
      x=rd.random()
      U= x0 + gam * math.tan(math.pi * (x-0.5))
      C.append(U)
  return C
```

```python
def distriangular(a,b,c,n):
  C=[]
  d1=(b-a)
  d2=(c-a)
  d3=(c-b)

  for i in range(n):
    x=rd.random()
    if x<=0:
      U = 0
      C.append(U)
    elif 0 < x < d1/d2:
      U = np.sqrt(d1 * d2 * x) + a
      C.append(U)

    elif d1/d2<x<1:
      k=-(x-1)
      U=-np.sqrt(k*d2*d3)+c
      C.append(U)
    elif x>=1:
      U=1
      C.append(U)


  return C
```

```python
G=distriangular(100,150,200,90000)

plt.plot(G)
```

```python
def dist_comp(n):
  C=[]
  for i in range(n):
    x=rd.random()
    if x<0:
      U = 0
      C.append(U)
    elif 0 <= x < 0.75:
      U = 1.25 - np.sqrt(1.5625 -2*x)
      C.append(U)

    elif 0.75<=x<1:
      U= 4 * x -2
      C.append(U)
    elif x>=1:
      U=1
      C.append(U)


  return C
```

# Método de aceptacion y rechazo
Generar una muestra de la variable con distribución
$$f(x)=20x(1-x)^{3}, \qquad 0 <x< 1 $$

## Teorema fundamenrtal de la simulacion

Simular $X \sim f(x)$ es equivalente a simular
$$(X,U) \sim u\{(x,u) : 0 < u < f(x) \}   $$

Esto se puede reescribir como:

$$ f(x) = \int \limits_{0}^{f(x)}  du$$

### Ejemplo:

Queremos generar un $x \sim f(x)$, suponiendo que sabemos como generar $g(x)$ y sabemos que:

$$\frac{f(y)}{g(y)} \leq c ;\quad  para \quad todo \quad y $$

## Metodo:

* Generar $y$ con distribución $g(x)$.
* Generar un numero aleatorio U
* si $U \leq \frac{f(y)}{g(y)}$ , hacer $x=y$.




```python
def beta(x):
  return 20*x*(1-x)**3
```

```python
def betag(g,n):


  C=[]

  for i in range(n):
    Y=rd.random()
    U=rd.random()
    h=beta(U)#unicamente en este caso
    if U<=h:
      U=Y
      C.append(Y)




  return  Y
```

```python


h=betag(1,200)
plt.hist(h, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')

# Agregar etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución normal estándar generada con aproximacion')

# Mostrar la gráfica
plt.show()
```

```python
def box(n):

  X1=[]
  X2=[]
  C=[]

  for i in range(n):
    u1=rd.random()
    u2=rd.random()
    x1 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    C.append(x1)
    x2 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)
    C.append(x2)
    X1.append(x1)
    X2.append(x2)

  return X1,X2,C
```

```python



h, g, H=box(50000)

plt.hist(h, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')

# Agregar etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución normal estándar generada con Box - Muller de x1')

# Mostrar la gráfica
plt.show()


plt.hist(g, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')

# Agregar etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución normal estándar generada con Box - Muller de x2')

# Mostrar la gráfica
plt.show()


plt.hist(H, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')

# Agregar etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución normal estándar generada con Box - Muller de x1 y X2')

# Mostrar la gráfica
plt.show()
```
