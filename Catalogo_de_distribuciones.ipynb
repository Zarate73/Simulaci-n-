```python
def distriangular(a,b,c,n):
  C=[]
  d1=(b-a)
  d2=(c-a)
  d3=(c-b)

  for i in range(n):
    x=rd.random()
    if x<=0:
      U = 0
      C.append(U)
    elif 0 < x < d1/d2:
      U = np.sqrt(d1 * d2 * x) + a
      C.append(U)

    elif d1/d2<x<1:
      k=-(x-1)
      U=-np.sqrt(k*d2*d3)+c
      C.append(U)
    elif x>=1:
      U=1
      C.append(U)


  return C
```

# Mtodo de aceptacion y rechazo
Generar una muestra de la variable con distribucin
$$f(x)=20x(1-x)^{3}, \qquad 0 <x< 1 $$

## Teorema fundamenrtal de la simulacion

Simular $X \sim f(x)$ es equivalente a simular
$$(X,U) \sim u\{(x,u) : 0 < u < f(x) \}   $$

Esto se puede reescribir como:

$$ f(x) = \int \limits_{0}^{f(x)}  du$$

### Ejemplo:

Queremos generar un $x \sim f(x)$, suponiendo que sabemos como generar $g(x)$ y sabemos que:

$$\frac{f(y)}{g(y)} \leq c ;\quad  para \quad todo \quad y $$

## Metodo:

* Generar $y$ con distribucin $g(x)$.
* Generar un numero aleatorio U
* si $U \leq \frac{f(y)}{g(y)}$ , hacer $x=y$.




```python
def beta(x):
  return 20*x*(1-x)**3
```

```python
def beta_acep(N):
    C=[] #Matriz vacia donde se almacenran los valores aceptados
    for i in range(N):
        y=rd.random() # Variable que sigue la distribucion g(x)
        u= np.random.uniform(0, 1) # Variable que sigue una distribucion uniforma
        if u<=beta(y)/beta(0.25):
            C.append(y)
    return C
```

```python
G=beta_acep(100000)
# Crear histograma para visualizar la distribucin
plt.hist(G, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')

# Agregar etiquetas y ttulo
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribucin beta generada con aceptacion y rechazo')
plt.grid(True)
# Graficar la funcin de densidad terica
x = np.linspace(0, 1)
y=beta(x)
plt.plot(x,y,'r-', lw=2,label='y = 20x(1-x)^3')
plt.legend()


# Mostrar la grfica
plt.show()
```

```python
def box(n):

  X1=[]
  X2=[]
  C=[]

  for i in range(n):
    u1=rd.random()
    u2=rd.random()
    x1 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    C.append(x1)
    x2 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)
    C.append(x2)
    X1.append(x1)
    X2.append(x2)

  return X1,X2,C
```

## Experimento

```python
def estimacion_pi(N):
  n=0
  K=0
  C=[]
  for i in range(N):
    x=rd.random()
    y=rd.random()
    if y <= np.sqrt(1-x**2):
      n+=1
      K=4*(n/(i+1))
      C.append(K)


    p=n/N
    estimate=4*p


  return C, estimate
```

Si bien ya generamos el experimento es momento de hacer la simulacin, lo que implica realizar el experimeto varias veces. Por lo que reescribimos el cdigo ya que nos interesa ahora la grfica de convergencia si no solo el ltimo dato.



```python
def est_pi(N):
  n=0
  for i in range(N):
    x=rd.random()
    y=rd.random()
    if y <= np.sqrt(1-x**2):
      n+=1

    estimate=4*(n/N)
  return  estimate
```

```python
def sim_pi(M):
  C=[]
  cont=0
  k=0
  while np.absolute(k-np.pi)>0.00001:
    k=est_pi(cont+1)
    C.append(k)
    cont +=1
    # Agregar una condicin de parada adicional para evitar un bucle infinito
    if cont > 1000000:
      print("Bucle infinito detectado. Saliendo...")
      break
  return C, cont, k
```

```python
k, j, L= sim_pi(1)
print("Iteraciones:",j)
print("Estimaciones:",L)
print(np.mean(k))


```
