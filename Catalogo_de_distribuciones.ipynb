```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python

def expo(lam,n):

  Mu=1/lam
  L=-Mu
  C=[]
  for i in range(n):

      x=rd.random()
      U=L*np.log10(x)
      C.append(U)


  return C
```

```python
G=expo(0.5,1000000)
print(np.transpose(G))
```

```python
def cdf(muestra):

  n=len(muestra)
  x=np.sort(muestra)
  y = np.arange(1, n + 1) / n
  return x,y
```

```python
X, Y=cdf(G)
plt.step(X,Y,"r")
```

```python

def Cauchy(gam,x0,n):
  C=[]
  for i in range(n):
      x=rd.random()
      U= x0 + gam * math.tan(math.pi * (x-0.5))
      C.append(U)
  return C
```

```python
N=Cauchy(1,0,4500)
plt.hist(N, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')

# Agregar etiquetas y ttulo
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribucin Cauchy generada con Transformada inversa')

# Mostrar la grfica
plt.show()
#print(np.transpose(N))
```

```python
def distriangular(a,b,c,n):
  C=[]
  d1=(b-a)
  d2=(c-a)
  d3=(c-b)

  for i in range(n):
    x=rd.random()
    if x<=0:
      U = 0
      C.append(U)
    elif 0 < x < d1/d2:
      U = np.sqrt(d1 * d2 * x) + a
      C.append(U)

    elif d1/d2<x<1:
      k=-(x-1)
      U=-np.sqrt(k*d2*d3)+c
      C.append(U)
    elif x>=1:
      U=1
      C.append(U)


  return C
```

```python
G=distriangular(100,150,200,90000)

plt.plot(G)
```

```python
import numpy as np
X, Y=cdf(G)
plt.step(X,Y,"r",)
plt.grid(True)
```

```python
def dist_comp(n):
  C=[]
  for i in range(n):
    x=rd.random()
    if x<0:
      U = 0
      C.append(U)
    elif 0 <= x < 0.75:
      U = 1.25 - np.sqrt(1.5625 -2*x)
      C.append(U)

    elif 0.75<=x<1:
      U= 4 * x -2
      C.append(U)
    elif x>=1:
      U=1
      C.append(U)


  return C
```

```python
Y=dist_comp(110000)
plt.plot(Y)
```

```python

plt.hist(Y, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')

# Agregar etiquetas y ttulo
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribucin compuesta generada con funcin inversa')

# Mostrar la grfica
plt.show()
```

```python
import numpy as np
X, Y=cdf(Y)
plt.step(X,Y,"r",)
plt.grid(True)
```

```python
def dist_aprox(n):
  C=[]
  h=0.1975
  for i in range(n):
    x=rd.random()
    j=x**0.135
    k=(1-x)**0.135
    U= (j-k)/h
    C.append(U)


  return C
```

```python
M=dist_aprox(11000000)
plt.plot(M)
```

```python
plt.hist(M, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')

# Agregar etiquetas y ttulo
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribucin normal estndar generada con aproximacion')

# Mostrar la grfica
plt.show()
```

```python
def dist_We(alpha,n):

  C=[]
  j=1/alpha
  for i in range(n):
    x=rd.random()
    U=(-np.log10(x))**j
    C.append(U)

  return C
```

```python
def dist_Gama(n):
#Entre 0 y 1
  C=[]

  for i in range(n):
    x=rd.random()
    U=(-np.log10(-np.log10(x)))
    C.append(U)

  return C
```

```python
Gma= dist_Gama(20000)
plt.hist(Gma, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')

# Agregar etiquetas y ttulo
plt.xlabel('Valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribucin Gama generada con Transformada inversa')

# Mostrar la grfica
plt.show()
```

```python
def dist_L(n):#Entre 0 y 1
#Entre 0 y 1
  C=[]

  for i in range(n):
    x=rd.random()
    U=-np.log10((1-x)/x)
    C.append(U)

  return C
```
