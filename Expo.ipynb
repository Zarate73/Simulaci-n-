```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

```python

def expo(lam,n):

  Mu=1/lam
  L=-Mu
  C=[]
  for i in range(n):

      x=rd.random()
      U=L*np.log10(x)
      C.append(U)


  return C
```

```python
G=expo(0.5,10)
print(np.transpose(G))
```

```python
media=np.mean(G)
print(media)
mu=1/0.5
print(mu)
```

```python
plt.hist(G,19)
```

```python
def cdf(muestra):

  n=len(muestra)
  x=np.sort(muestra)
  y = np.arange(1, n + 1) / n
  return x,y
```

```python
X, Y=cdf(G)
plt.step(X,Y,"r")
```
